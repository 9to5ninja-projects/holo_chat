## üß† **CREATE DIGITAL BRAIN WITH LOCAL LLM**

```python
# üß† CREATE DIGITAL BRAIN WITH LOCAL LLM
print("üß† CREATING DIGITAL BRAIN WITH LOCAL LLM")
print("=" * 45)

if not SYSTEM_READY:
    print("‚ùå System not ready. Please fix setup first.")
    brain = None
else:
    # === CONFIGURE LOCAL LLM ===
    print("\nü§ñ Configuring Local LLM...")
    
    # You can modify these settings based on your setup
    LLM_CONFIG = {
        # Ollama settings (if available)
        'ollama_model': 'llama2',  # or 'mistral', 'codellama', etc.
        
        # Transformers settings
        'transformers_model': 'microsoft/DialoGPT-medium',
        
        # LlamaCpp settings (provide path to your GGUF model)
        'llamacpp_model_path': None,  # e.g., '/path/to/llama-2-7b-chat.gguf'
        
        # Preferred LLM type ('auto', 'ollama', 'transformers', 'llamacpp')
        'preferred_llm': 'auto'
    }
    
    # Create LLM interface
    try:
        if LLM_CONFIG['preferred_llm'] == 'ollama':
            llm_interface = LocalLLMFactory.create_ollama(
                model_name=LLM_CONFIG['ollama_model']
            )
            print(f"‚úÖ Using Ollama with model: {LLM_CONFIG['ollama_model']}")
            
        elif LLM_CONFIG['preferred_llm'] == 'transformers':
            llm_interface = LocalLLMFactory.create_transformers(
                model_name=LLM_CONFIG['transformers_model']
            )
            print(f"‚úÖ Using Transformers with model: {LLM_CONFIG['transformers_model']}")
            
        elif LLM_CONFIG['preferred_llm'] == 'llamacpp':
            if LLM_CONFIG['llamacpp_model_path']:
                llm_interface = LocalLLMFactory.create_llamacpp(
                    model_path=LLM_CONFIG['llamacpp_model_path']
                )
                print(f"‚úÖ Using LlamaCpp with model: {LLM_CONFIG['llamacpp_model_path']}")
            else:
                print("‚ùå LlamaCpp model path not specified")
                raise ValueError("Please provide llamacpp_model_path")
                
        else:  # auto-detect
            llm_interface = LocalLLMFactory.auto_detect_and_create(
                model_name=LLM_CONFIG['ollama_model']
            )
            print("‚úÖ Auto-detected and configured local LLM")
            
    except Exception as e:
        print(f"‚ùå Error setting up local LLM: {e}")
        print("\nüîß Troubleshooting:")
        print("1. For Ollama: Make sure 'ollama serve' is running and model is pulled")
        print("2. For Transformers: Ensure sufficient RAM/VRAM for model")
        print("3. For LlamaCpp: Provide valid path to GGUF model file")
        llm_interface = None
    
    # === CREATE DIGITAL BRAIN ===
    if llm_interface:
        print("\nüß† Creating Digital Brain...")
        
        # Configure XP Core for consciousness
        consciousness_config = UnifiedXPConfig(
            embedding_dim=384,      # Rich semantic representations
            hrr_dim=512,           # Holographic memory capacity
            decay_half_life=72.0,  # 3-day memory persistence
            k_neighbors=12,        # Rich associative context
            importance_boost_factor=1.5,  # Boost important memories
            consolidation_threshold=0.7  # Strong consolidation
        )
        
        # Create the digital brain
        brain = DigitalBrain(
            name="Lumina",
            config=consciousness_config,
            llm_interface=llm_interface
        )
        
        print(f"‚úÖ Digital Brain 'Lumina' created successfully!")
        print(f"   Memory System: {type(brain.memory_core).__name__}")
        print(f"   Language Model: {type(brain.language_model).__name__}")
        print(f"   Birth Time: {datetime.fromtimestamp(brain.birth_time)}")
        
    else:
        print("‚ùå Cannot create Digital Brain without LLM interface")
        brain = None

print(f"\nüéØ Digital Brain Status: {'‚úÖ Ready' if brain else '‚ùå Not Ready'}")
```
